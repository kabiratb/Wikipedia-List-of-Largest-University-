```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 12))
sns.barplot(y='Country', x='Enrollment', data=df, errorbar=None, hue='Country', palette='viridis')
plt.title('Number of Universities per Country with Enrollment')
plt.xlabel('Enrollment')
plt.ylabel('Country')
plt.show()
```

```python
#University by Year founded
df['Founded'] = df['Founded'].apply(pd.to_numeric)

plt.figure(figsize=(12, 6))
sns.histplot(df['Founded'], bins=30, kde=True)
plt.title('Distribution of Universities by Founding Year')
plt.xlabel('Year Founded')
plt.ylabel('Number of Universities')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Type', data=df, hue='Type', palette='Blues_d')
plt.title('Type of Universities (Public vs. Private)')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
#To 10 Universities By Enrollment
df['Enrollment'] = df['Enrollment'].apply(pd.to_numeric)

top_enrollment = df.nlargest(10, 'Enrollment')

# Create a column to indicate the top university
top_enrollment['Top'] = top_enrollment['Enrollment'] == top_enrollment['Enrollment'].max()

# Define colors for 'Top' and 'Not Top'
palette = {True: 'red', False: 'blue'}

plt.figure(figsize=(10, 6))
sns.barplot(x='Enrollment', y='University', data=top_enrollment, hue='Top', palette=palette)
plt.title('Top 10 Universities by Enrollment')
plt.xlabel('Enrollment')
plt.ylabel('University')
plt.legend(title='Top University')
plt.show()
```

```python
#Enrollment By Year the University Was Founded
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Founded', y='Enrollment', data=df)
plt.title('Enrollment vs. Year Founded')
plt.xlabel('Year Founded')
plt.ylabel('Enrollment')
plt.show()

```

```python
#Average Enrollment with the Type of Unversity
avg_enrollment = df.groupby('Type')['Enrollment'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Type', y='Enrollment', data=avg_enrollment, hue='Type', palette='Blues_d')
plt.title('Average Enrollment by Type of University')
plt.xlabel('Type')
plt.ylabel('Average Enrollment')
plt.show()
```
